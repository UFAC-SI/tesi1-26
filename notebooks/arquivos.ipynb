```python
# Criando um aquivo .txt
especies = ['setosa', 'virginica', 'versicolor']
with open('especies.txt', 'w') as arq:
    for especie in especies:
        arq.write(especie+'\n')
print('Arquivo gravado com sucesso!')
```

```python
# Lendo um arquivo .txt
with open('especies.txt', 'r') as arq:
    #linhas = arq.readlines()
    for linha in arq:
        print(linha, end='')
```

```python
# Adicionar uma linha ao arquivo
with open('especies.txt', 'a') as arq:
    arq.write('rosa\n')
print('Linha adicionada!')
```

```python
# Busca e remoção de uma linha no arquivo .txt
with open('especies.txt', 'r') as leitura:
    linhas = []
    for linha in leitura:
        linhas.append(linha.strip())
    #removido = 'virginica'
    #linhas.remove(removido)
    #for linha in linhas:
    #    escrita.write(linha+'\n')
    
with open('especies.txt', 'w') as escrita:
    for linha in linhas:
        if linha != 'virginica':
            escrita.write(linha+'\n')
print('Arquivo recriado!')
```

```python
# Criando um arquivo .csv
colunas = ['id', 'tamanho', 'especie']
linhas = [[1,2.5,'versicolor'],[2,3.5,'virginica'],[3,4.5,'setosa']]
#print(','.join(colunas)+'\n')
with open('especies.csv', 'w') as arq:
    arq.write(','.join(colunas)+'\n')
    for linha in linhas:
        arq.write(','.join([str(i) for i in linha])+'\n')
```

```python
# Leitura com csv
import csv
with open('especies.csv', 'r') as arq:
    cursor = csv.DictReader(arq)
    #print(','.join(next(cursor).keys()))
    print('\t'.join(cursor.fieldnames)) #percorrendo os dados com iterador
    for linha in cursor:
        print(linha['id'], linha['tamanho'], linha['especie'], sep='\t')
```

```python
# Inserindo uma linha no csv
nova_linha = [5, 5.6, 'rosa']
#print(nova_linha)
with open('especies.csv', 'a') as arq:
    arq.write(','.join([str(j) for j in nova_linha])+'\n')
```

```python
import csv, json
with open('especies.csv', 'r') as arq:
    cursor = csv.DictReader(arq)
    #print(next(cursor))
    print(','.join(cursor.fieldnames))
    lista = list(cursor)
    #print(lista)
    for linha in lista:
        print(','.join(linha.values()))   
```

```python
# Criando um arquivo json
with open('especies.json', 'w') as arq:
    json.dump(lista, arq, indent=4, ensure_ascii=False)
print('Arquivo json salvo com sucesso!')
```

```python
# Lendo um arquivo json
with open('especies.json', 'r') as arq:
    lista = json.load(arq)
    print(','.join(lista[0].keys()))
    for linha in lista:
        print(','.join(linha.values()))
```

```python
# Buscar informações no json
chave = 'especie'
valor = 'rosa'
resultados = [linha for linha in lista if linha[chave] == valor]
print(resultados)
with open('resultados_especie.json', 'w') as arq:
    json.dump(resultados, arq, indent=4, ensure_ascii=False)
```

```python
# Salvando as especies que não são rosas
resultados2 = [linha for linha in lista if linha[chave] != valor]
with open('resultados_especie2.json', 'w') as arq:
    json.dump(resultados2, arq, indent=4, ensure_ascii=False)
```
