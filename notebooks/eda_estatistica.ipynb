## Análise Exploratória de dados com gráficos e estatística

#### Informações básicas

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.info()
```

```python
# Estatística Descritiva dos dados númericos
tips.describe()
```

```python
# Estatística Descritiva dos dados categóricos
tips.mode()
```

```python
# Média de uma coluna específica
tips['tip'].mean()
```

```python
# Agrupando a média em um atributo categórico
tips.groupby('day', observed=True)['tip'].mean()
```

```python
# Agrupando várias medidas em um atributo categórico
tips.groupby('day', observed=True)['tip'].agg(['mean', 'median', 'std'])
```

#### Gráficos

```python
# Histograma
import matplotlib.pyplot as plt
sns.histplot(data=tips, x='tip', bins=10)
plt.title('Distribuição das Gorjetas')
plt.xlabel('Valor das Gorjetas')
plt.ylabel('Frequência')
plt.show()
```

```python
# Histograma por categoria
import matplotlib.pyplot as plt
sns.histplot(data=tips, x='tip', hue='sex', multiple='stack',bins=10)
plt.title('Distribuição das Gorjetas por Gênero')
plt.xlabel('Valor das Gorjetas')
plt.ylabel('Frequência')
plt.legend(title='Gênero', labels=['Feminino', 'Masculino'])
plt.savefig('histograma_gorjetas_por_genero.pdf')
plt.show()
```

```python
# Boxplot

sns.boxplot(data=tips, y='total_bill', width=.2)
plt.show()
```

```python
# Boxplot por categoria
sns.boxplot(data=tips, x='day', y='total_bill', width=.2, showmeans=True)
plt.show()
```

```python
import numpy as np
sns.barplot(data=tips, x='day', y='tip', hue='day', errorbar=None, estimator=np.sum)
plt.show()
```
