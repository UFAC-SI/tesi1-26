## Identificando dados ausentes

```python
import matplotlib.pyplot as plt
import numpy as np

copia = df.copy()
indices = np.random.choice(copia.index, size=100, replace=False)
copia.loc[indices, 'tip'] = np.nan
sns.heatmap(copia.isnull())
plt.show()
removidos = copia.dropna()
print(removidos.shape)
```

```python
print(df['tip'].median())
print(copia['tip'].median())
mediana = copia['tip'].median()
print(copia.shape)
preenchido = copia.copy()
print(preenchido.shape)
preenchido['tip'] = preenchido['tip'].fillna(0)
print(preenchido['tip'].median())
```
